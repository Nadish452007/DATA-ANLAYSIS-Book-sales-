```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Books_Data_Clean.csv")
df
```

```python
df.head()
```

```python
df.describe()
```

```python
df=df[df["Publishing Year"]>1900]
df
```

```python


df.isna().sum()
```

```python
df.dropna(subset="Book Name",inplace=True)
```

```python
df.nunique()
```

```python


plt.hist(df["Publishing Year"])
plt.xlabel("YEAR")
plt.ylabel("frequencys")
plt.title("Graph for publishing year",color="red")
plt.show()
```

```python
colors=["Orange","blue","green"]
h=df["genre"].value_counts()
h.plot(kind="bar",color=colors)
plt.xlabel("genre")
plt.ylabel("number of books")
plt.title("number of books in each genre")
plt.show()
```

```python
df.groupby(["Author","gross sales","Book_average_rating"])["sale price"].sum().sort_values(ascending=True).head()
```

```python
sns.boxplot(x="genre",y="Book_ratings_count",data=df, color="#1E93AB")
plt.xlabel("genre")
plt.ylabel("counts for the books")
plt.title("number of books in each rating_count")
plt.show()
```

```python

plt.scatter(df["sale price"],df["units sold"])
plt.ylabel("UNIT SALE")
plt.xlabel("SALE PRICE")
plt.show()
```

```python
h=df["language_code"].value_counts().sort_values(ascending=True)
```

```python
colors=["pink","yellow","skyblue","red","skyblue"]
plt.pie(h,labels=h.index,autopct="%1.1f",colors=colors,shadow=True)
plt.title("Graph for language in books")
plt.legend()
plt.show()
```
